```python
graph = {
    'a': ['c'],
    'b': ['c', 'e'],
    'c': ['a', 'b', 'd', 'e'],
    'd': ['c'],
    'e': ['b', 'c'],
    'f': []
}
def graph_connected(graph, seen_node=None, start=None):
    if seen_node is None:
        seen_node = set()
        nodes = list(graph.keys())  # List of all the graph keys
        if not start:  # If start is not given
            start = nodes[0]  # Start from the first node in the graph
        seen_node.add(start)   
    # Recursively visit each node
    for othernodes in graph[start]:  # For the adjacent nodes of start
        if othernodes not in seen_node:
            seen_node.add(othernodes)
            graph_connected(graph, seen_node, othernodes)  # Recur for unvisited nodes
    
    # If all nodes have been visited, the graph is connected
    return len(seen_node) == len(graph)
# Checking if the graph is connected
conn = graph_connected(graph)
if conn:
    print("The graph is connected")
else:
    print("The graph is not connected")
```
