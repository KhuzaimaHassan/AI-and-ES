```python
graphA= {
  'A' : ['B','E'],
  'B' : ['F'],
  'C' : ['G'],
  'D' : ['E','H'],
  'E' : ['A','D','H'],
  'F' : ['B','G','I','J'],
  'G' : ['C','F','J'],
  'H' : ['D','E','I'],
  'I' : ['F','H'],
  'J' : ['F','G']
}
graphB={
    'A':['B','C','E'],
    'B':['A','D','E'],
    'C':['A','F','G'],
    'D':['B','E'],
    'E':['A','B','D'],
    'F':['C'],
    'G':['C']
}
from collections import deque
def bfs_shortest_path(graph, start, end):
  queue = deque( )
# Enqueue the start node and mark it visited
  queue.append( ( start, [start])) 
  visited = set([start])
  while queue:
    node, path = queue.popleft()
    if node == end:
      return path
    for neighbour in graph[node]:
      if neighbour not in visited:
        visited.add(neighbour)
        queue.append((neighbour, path + [neighbour]))

print(bfs_shortest_path(graphA,'G','D'))
print(bfs_shortest_path(graphB,'G','D'))
```

```python
graphC={
    '1':['2','3','4'],
    '2':['1','3','4'],
    '3':['1','2','4'],
    '4':['1','2','3','5'],
    '5':['4','6','7','8'],
    '6':['5','7','8'],
    '7':['5','6','8'],
    '8':['5','6','7']
}

#Apply BFS to find to every possible node present in graph.Starting from 1.
def bfs_connected_component(graph, start) :
  explored =[]
  queue= [start]
# keep looping until there are nodes stilt
  while queue:
    node = queue. pop(0)
    if node not in explored:
       explored. append (node)
       neighbours = graph[node]
# add neighbours of node to queue
       for neighbour in neighbours:
           queue. append (neighbour)
  return explored
print('Every Possible Node starting from 1: ',bfs_connected_component(graphC,'1'))

#Find all paths between 1 & 6.
def all_paths(graph, start, goal):
  queue=[(start, [start])]
  while queue:
    (vertex,path)=queue.pop(0)
    for next in set(graph[vertex]) - set(path):
      if next == goal:
        yield path + [next]
      else:
        queue.append((next, path + [next]))
print('\nAll Paths between 1 and 6: ',list(all_paths(graphC,'1','6')))

#Find shortest path between 1 & 6.
from collections import deque
def bfs_shortest_path(graph, start, end):
  queue = deque( )
# Enqueue the start node and mark it visited
  queue.append( ( start, [start])) 
  visited = set([start])
  while queue:
    node, path = queue.popleft()
    if node == end:
      return path
    for neighbour in graph[node]:
      if neighbour not in visited:
        visited.add(neighbour)
        queue.append((neighbour, path + [neighbour]))

print('\nShortest Path between 1 and 6: ',bfs_shortest_path(graphC,'1','6'))
```
